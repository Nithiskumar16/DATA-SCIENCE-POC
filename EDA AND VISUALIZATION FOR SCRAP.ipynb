**DROP  COLUMNS**

```python
dd= data.copy()
```

```python
dd.drop(['stroke','highwaympg', 'peakrpm','wheelbase', 'ID'],axis = 1,inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dd['name'] = le.fit_transform(dd['name'])
dd['aspiration']= le.fit_transform(dd['aspiration'])
dd['drivewheels']= le.fit_transform(dd['drivewheels'])
dd['enginelocation']= le.fit_transform(dd['enginelocation'])
dd['enginetype']= le.fit_transform(dd['enginetype'])
dd['cylindernumber']= le.fit_transform(dd['cylindernumber'])
dd['fuelsystem']= le.fit_transform(dd['fuelsystem'])
dd['carbody']= le.fit_transform(dd['carbody'])
```

```python
dd['fueltypes']= le.fit_transform(dd['fueltypes'])
```

```python
dd.replace({'two':2,'four':4 }, inplace = True)
```
