**FEATURE SCALING AND MODEL BUILDING **

```python
from sklearn import preprocessing
stan_rfm=preprocessing.scale(RFM)
```

**CLUSTERING BY ELBOW METHOD**

```python
from sklearn.cluster import KMeans
inertia=[]
for i in np.arange(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(stan_rfm)
  inertia.append(kmeans.inertia_)
```

```python
plt.plot(inertia,marker ="*")
```

```python
kmeans =KMeans(n_clusters =3)
kmeans.fit(stan_rfm)
RFM['cluster']=(kmeans.labels_+1)
```

```python
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4
```

```python
def FScore(x,t,d):
    if x <= d[t][0.25]:
        return 4
    elif x <= d[t][0.50]:
        return 3
    elif x <= d[t][0.75]:
        return 2
    else:
        return 1
```

```python
def MScore(x,s,d):
    if x <= d[s][0.25]:
        return 4
    elif x <= d[s][0.50]:
        return 3
    elif x <= d[s][0.75]:
        return 2
    else:
        return 1
```

```python
SCRFM['R_Quartile'] = RFM['Recency'].apply(RScore,args=('Recency',quantile))
SCRFM['F_Quartile'] = RFM['Frequency'].apply(FScore,args=('Frequency',quantile))
SCRFM['M_Quartile'] = RFM['Monetary'].apply(MScore,args=('Monetary',quantile))
```
